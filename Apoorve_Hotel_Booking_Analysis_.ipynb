# **Data Cleaning**
<font color='red'>The process of Data cleaning starts as folllowing :

<font color='red'>1. Optimize dataset while removing the duplicates.

<font color='red'>2. Finding the null values in dataset , either assign 0 values to that column or remove that columns if that column are not important for analysis .

```python
# checking for null values

null = pd.DataFrame({'Null_values' : df1.isna().sum(), 'Percentage_null_values' : (df1.isna().sum()) / (df.shape[0]) * (100)})
null
```

```python
# Creating heatmap for overview of missing values in whole dataframe
plt.rcParams["figure.figsize"] = (15, 7.5)
sns.heatmap(df1.isnull(), cbar=False,cmap='flare',yticklabels=False)
plt.xlabel("Name Of Columns", fontsize=18)
plt.title('Finding missing values in columns',fontsize=25)
plt.show()
```

```python
# filling null values with zero

df1.fillna(0, inplace = True)
```

```python
# adults, babies and children cant be zero at same time, so dropping the rows having all these zero at same time

filter = (df1.children == 0) & (df1.adults == 0) & (df1.babies == 0)
df1[filter]
```

```python
df1 = df1[~filter]
df1
```

```python
# Finally check the dataframe for cinsisrtency in records
plt.rcParams["figure.figsize"] = (15, 7.5)
sns.heatmap(df1.isnull(), cbar=False,cmap='flare',yticklabels=False)
plt.xlabel("Name Of Columns", fontsize=18)
plt.title('Finding missing values in columns',fontsize=25)
plt.show()
```
